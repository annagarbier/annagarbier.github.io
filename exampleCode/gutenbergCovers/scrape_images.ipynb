```python
from requests_html import HTMLSession
from time import sleep
import random as rand

session = HTMLSession()
```

```python
import pandas as pd

gutenberg_df = pd.DataFrame(columns=['uid', 'cover_src'])
gutenberg_df
```

```python
responses = []
errors = []
for i in range(0,60472):
    if (i % 25) == 0:
        try:
            response = session.get('https://www.gutenberg.org/ebooks/' + str(i))

            # Book id
            uid = i

            # Cover art
            cover_element = response.html.find("img.cover-art")
            cover_src = cover_element[0].attrs.get("src")
            responses.append((str(i) + ": " + cover_src + " "))

            # Update dataframe
            gutenberg_df = gutenberg_df.append({'uid': uid,
                'cover_src': cover_src}, ignore_index=True)

            # Rand sleep before next session.get
            time_delay = rand.randrange(0, 1)
            sleep(time_delay)

        except Exception as e:
            errors.append(e)

print("done")
```

```python
print("responses: " + str(len(responses)))
print("errors: " + str(len(errors)))
```

```python
# Download images

import urllib.request

for index, row in gutenberg_df.iterrows():
    # Full image url
    url = row['cover_src']
    
    # File name only
    file_name = url.split('/')[-1]
    
    # Print progress
    print("downloading " + url + " as " + file_name)
    
    # Download
    try:
        urllib.request.urlretrieve(url, 'scraped_images_900/' + file_name)
    except Exception as e:
        print(e)
```
